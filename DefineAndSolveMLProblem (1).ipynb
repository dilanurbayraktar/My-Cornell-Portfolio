```python
import scipy.stats as stats

df = pd.read_csv(airbnbDataSet_filename, header=0)

# Step 1: Check if the label needs to be reformatted. It is a float, so no need to reformat.
print(df.dtypes['price'])
print(df['price'].unique()[:16])

# Step 2: To select features, drop any columns that would not serve as features

list_drop = ['name', 'description', 'neighborhood_overview', 'host_name', 'host_location', 'host_about']
df.drop(columns=list_drop, inplace=True)

features = df.columns.tolist()

print(features)

# Step 3: Handle Outliers
df['label_price'] = stats.mstats.winsorize(df['price'], limits=[0.01, 0.01])

print(df.head())

# Step 4: Handle Missing Data
nan_count = df.isna().sum()
print(nan_count) # There are some missing data

nan_detected = nan_count > 0

is_int_or_float = df.dtypes.apply(lambda x: x in [np.int64, np.float64])
print(is_int_or_float)

to_impute = nan_detected & is_int_or_float
print(to_impute)

print(df.columns[to_impute])

to_impute_selected = ['host_response_rate', 'host_acceptance_rate','bedrooms', 'beds']

for clmn in to_impute_selected:
    df[f'{clmn}_na'] = df[clmn].isna()

#Replace missing values with mean values
for column in to_impute_selected:
    mean = df[column].mean()
    df[column].fillna(value = mean, inplace = True)

for colname in to_impute_selected:
    print("{} missing values count :{}".format(colname, np.sum(df[colname].isnull(), axis = 0)))

# Step 5: One-Hot Encoding
for colname in features:
    if df.dtypes[colname] == object:
        print("{} type: {}".format(colname, df.dtypes[colname], axis = 0))

# I dentified that 'neighbourhood_group_cleansed', 'room_type', and 'amenities' features are strings
print(df['neighbourhood_group_cleansed'].unique())
print(df['room_type'].unique())
print(df['amenities'].unique())

# None of them has NaN as a value

# Identifying potential candidates for One-Hot Encoding
string_val = df.select_dtypes(include=['object']).columns
unique_num = df[string_val].nunique()
print(unique_num)

encode_column = unique_num[unique_num < 20].index.tolist()
for column in encode_column:
    print(f"\nUnique values in column '{column}':")
    print(df[column].unique())


# Decided that One-Hot Encoding should be applied on 'neighbourhood_group_cleansed' and 'room_type'
# So, complete One-Hot Encoding
df_neighbourhood_group_cleansed = pd.get_dummies(df['neighbourhood_group_cleansed'], prefix='neighbourhood_group_cleansed')
df = pd.concat([df, df_neighbourhood_group_cleansed], axis=1)
df.drop('neighbourhood_group_cleansed', axis=1, inplace=True)

df_room_type = pd.get_dummies(df['room_type'], prefix='room_type')
df = pd.concat([df, df_room_type], axis=1)
df.drop('room_type', axis=1, inplace=True)

# Drop Amenities as it's going to make the model more complex to work with
df.drop('amenities', axis=1, inplace=True)


# Step 4: Exploring the data
corr_matrix = round(df.corr(),5)

# Extract label_price correlation
corrs = corr_matrix['label_price']

# Sort the correlaction
corrs_sorted = corrs.sort_values(ascending=False)

# Identify the top two features, excluding label_price and price
top_two_corr = corrs_sorted.index[2:4].tolist()

# The top two correlated features are 'acommodates' and 'bedrooms'

# Plotting
df_corrs = df[['label_price'] + top_two_corr]

sns.pairplot(df_corrs, kind='kde', diag_kind='auto')

plt.show()

df.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
df.drop('price', axis=1)
# Step 1: Create labeled examples from the dataset
X = df.drop('label_price', axis=1)
y = df['label_price']

# Step 2: Create training and test data sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 1234)
```

```python
# Random Forest

print('Begin RF Implementation...')

rf_model = RandomForestRegressor(max_depth=32, n_estimators=300)

rf_model.fit(X_train, y_train)

print('End')

# 1. Use the fitted model to make predictions on the test data
y_rf_pred = rf_model.predict(X_test)


# 2. Compute the RMSE 
rf_rmse = mean_squared_error(y_test, y_rf_pred, squared=False)


# 3. Compute the R2 score 
rf_r2 = r2_score(y_test, y_rf_pred)


print('[RF] Root Mean Squared Error: {0}'.format(rf_rmse))
print('[RF] R2: {0}'.format(rf_r2))    
```

```python
# It looks like the model performed very well as the RMSE is low and R2 is almost 1!
```

```python
# Gradient Boost

from sklearn.ensemble import GradientBoostingRegressor

print('Begin GBDT Implementation...')

gbdt_model = GradientBoostingRegressor(max_depth=2, n_estimators=300)

gbdt_model.fit(X_train, y_train)

print('End')

# 1. Use the fitted model to make predictions on the test data
y_gbdt_pred = gbdt_model.predict(X_test)


# 2. Compute the RMSE 
gbdt_rmse = mean_squared_error(y_test, y_gbdt_pred, squared=False)


# 3. Compute the R2 score 
gbdt_r2 = r2_score(y_test, y_gbdt_pred)

print('[GBDT] Root Mean Squared Error: {0}'.format(gbdt_rmse))
print('[GBDT] R2: {0}'.format(gbdt_r2))                 
```

```python
# The R2 for GBDT was also close to 1, but the RMSE was too high, so not as good as Random Forest!
```
